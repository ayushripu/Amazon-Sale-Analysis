# To identify the most common order value range.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Amazon Sale Report.csv")
```

```python
plt.savefig("charts_fixed.png", dpi=300, bbox_inches='tight')
```

```python
import os
os.makedirs("images", exist_ok=True)

plt.hist(df['Amount'], bins=50)

plt.xlabel("Order Amount")
plt.ylabel("Number of Orders")
plt.title("Order Amount Distribution")

plt.savefig("images/Most_Order_Range.png", dpi=300, bbox_inches='tight')

plt.show()
```
